```python
import numpy as np

class agent:
  def __init__(self,gamma=0.99):
    self.v={}                  # value function
    self.sumspace=[i for i in range(4,22)]
    self.deal_space=[i+1 for i in range(10)]
    self.acespace=[False,True]
    self.actionspace=[0,1]  # stick or hit
    self.states_visited={}  # first visit
    self.memory=[]      # memory contains the state,reward tuple from each state in an episode
    self.statespace=[]
    self.returns={}     # state particular return
    self.gamma=gamma    
    self.init_vals()
  
  def init_vals(self):
    for s in self.sumspace:
      for c in self.deal_space:
        for ishita in self.acespace:
          self.v[(s,c,ishita)]=0
          self.returns[(s,c,ishita)]=[]         # returns from episodes over period of time
          self.states_visited[(s,c,ishita)]=0   # binary whether state is visited or not
          self.statespace.append((s,c,ishita))   # statespace

  
  def policy(self,state):
    # state consist of players total , dealer card , usable ace
    total,_,_=state
    action=0 if total>=20 else 1
    return action
  
  def update_v(self):
    for (idt,k) in enumerate(self.memory):
      state,_=k[0]
      G=0
      if self.states_visited[state]==0:
        self.states_visited[state]+=1
        discount=1
        for (t,l) in enumerate(self.memory[idt:]):
          _,reward=l[0]
          G+=reward*discount
          discount*= self.gamma
        self.returns[state].append(G)


    for st in self.memory:
      state,_=st[0]
      self.v[state]=np.mean(self.returns[state])

    for state in self.statespace:
      self.states_visited[state]=0       #reinitializing the states visited to 0 for different episode

    self.memory=[]
    
```

```python
Agent=agent()
```
